```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python

# CSVファイルを読み込む
input_file = "E:\実験\データセット\データセット2/2.次元圧縮後\主成分分析\主成分分析3（名詞数抜き）_処理前.csv"  # 入力ファイル名を適切に変更してください
df = pd.read_csv(input_file)

# データの前処理
# 数値データのみを選択（文字列や日付などは除外）
numeric_columns = df.select_dtypes(include=[np.number]).columns
X = df[numeric_columns]

# データの標準化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCAの実行（1次元に削減）
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)

# 結果をデータフレームに変換
df_result = pd.DataFrame(X_pca, columns=['PCA_1'])

# 元のデータフレームのインデックスを保持
df_result.index = df.index

# 結果をCSVファイルに出力
output_file = 'output_pca_result.csv'  # 出力ファイル名を適切に変更してください
df_result.to_csv(output_file, index=True)

print(f"PCA結果を {output_file} に出力しました。")
```
