```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python

# CSVファイルの読み込み
# 'your_data.csv'を実際のCSVファイル名に置き換えてください
df = pd.read_csv("E:\実験\データセット/3.VIF処理\VIF4処理\データセット（視線情報+やさにちチェッカー+意味情報）.csv")
```

```python

# 相関係数の計算
correlation = df.corr()

# 相関係数のヒートマップ
plt.figure(figsize=(20, 16))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Heatmap')
plt.tight_layout()
#plt.savefig('correlation_heatmap.png')
plt.savefig('相関行列_VIF処理後4.png')
plt.close()
```

```python

# VIFの計算
def calculate_vif(X):
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif

# 全ての数値列を使用してVIFを計算
X = df.select_dtypes(include=[np.number])
vif_data = calculate_vif(X)

# VIFの棒グラフ
plt.figure(figsize=(10, 6))
sns.barplot(x='Variable', y='VIF', data=vif_data)
plt.title('Variance Inflation Factor (VIF)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('vif_VIF処理後4.png')
plt.close()

print("相関係数とVIFの可視化が完了しました。")
print("生成された画像ファイル:")
print("- correlation_heatmap.png")
print("- vif_barplot.png")

# VIF値の表示
print("\nVIF値:")
print(vif_data)
```
