```python
import pandas as pd

# TSVファイルを読み込む
tsv_file = "E:/test.tsv"  # ここにTSVファイルのパスを入力してください
# ファイルのエンコーディングを確認し、必要に応じて変更してください
encoding = 'utf-8'  # 'shift-jis' など他のエンコーディングも試してみてください
df = pd.read_csv(tsv_file, sep='\t', encoding=encoding)

# CSVファイルに出力する
csv_file = 'aaa_文章1.csv'  # 出力するCSVファイルのパスを入力してください
df.to_csv(csv_file, index=False, encoding='utf-8-sig')

print(f'TSVファイルをCSVファイルとして{csv_file}に保存しました。')
```
