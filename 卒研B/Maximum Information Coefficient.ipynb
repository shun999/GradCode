```python
import pandas as pd
import numpy as np
# import seaborn as sns
import matplotlib.pyplot as plt
from minepy import MINE
```

```python

# Function to compute maximum information coefficient (MIC)
def calculate_mic(df):
    """
    Calculate Maximum Information Coefficient (MIC) for all pairs of variables in the DataFrame.
    
    :param df: DataFrame containing the dataset
    :return: DataFrame of MIC scores
    """
    columns = df.columns
    mic_matrix = np.zeros((len(columns), len(columns)))

    mine = MINE()

    for i, col1 in enumerate(columns):
        for j, col2 in enumerate(columns):
            if i == j:
                mic_matrix[i, j] = 1  # Self-correlation is always 1
            else:
                temp_df = df[[col1, col2]].dropna()
                x = temp_df[col1].values
                y = temp_df[col2].values
                mine.compute_score(x, y)
                mic_matrix[i, j] = mine.mic()

    mic_df = pd.DataFrame(mic_matrix, index=columns, columns=columns)
    return mic_df

# Plotting maximum information coefficient
def plot_mic(mic_matrix):
    plt.figure(figsize=(10, 8))
    sns.heatmap(mic_matrix, annot=True, fmt='.2f', cmap='viridis', cbar=True)
    plt.title('Maximum Information Coefficient (MIC) Heatmap')
    plt.show()

# Example Usage
# Replace 'your_dataset.csv' with the path to your dataset
# Ensure the dataset has been preprocessed (e.g., missing values, encoding categorical variables)
data = pd.read_csv('your_dataset.csv')

# Calculate MIC for the dataset (pairwise)
mic_matrix = calculate_mic(data)

# Plot the heatmap of MIC
plot_mic(mic_matrix)
```
