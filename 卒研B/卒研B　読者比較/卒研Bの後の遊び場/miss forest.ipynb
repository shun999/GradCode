```python
import pandas as pd
from missforest import MissForest
```

```python

# Load the dataset (adjust the file path if needed)
data = pd.read_csv("C:/Users/thisi/OneDrive/共有ファイル/卒研Bデータ/データセット原本_完全版.csv")

# Display initial missing value counts per column
print("Missing values before imputation:")
print(data.isnull().sum())

# Create the MissForest imputer instance
imputer = MissForest()

# Perform the imputation. This returns a NumPy array.
data_imputed = imputer.fit_transform(data)

# Convert the imputed array back to a DataFrame, retaining original column names and index
data_imputed_df = pd.DataFrame(data_imputed, columns=data.columns, index=data.index)

# Display missing value counts after imputation (should be zero if imputation was successful)
print("\nMissing values after imputation:")
print(data_imputed_df.isnull().sum())

# Save the imputed dataset to a new CSV file
data_imputed_df.to_csv('データセット原本_imputed.csv', index=False)

print("\nImputation completed. The imputed dataset has been saved as 'データセット原本_imputed.csv'.")
```

```python
# Load the dataset (adjust the file path if needed)
data = pd.read_excel("C:/Users/thisi/OneDrive/共有ファイル/卒研Bデータ/卒研後の遊び場/欠損値処理後.xlsx")

# Display initial missing value counts per column
print("Missing values before imputation:")
print(data.isnull().sum())

# Create the MissForest imputer instance
imputer = MissForest()

# Perform the imputation. This returns a NumPy array.
data_imputed = imputer.fit_transform(data)


# Convert the imputed array back to a DataFrame, retaining original column names and index
data_imputed_df = pd.DataFrame(data_imputed, columns=data.columns, index=data.index)

# Display missing value counts after imputation
print("\nMissing values after imputation:")
print(data_imputed_df.isnull().sum())


# Save the imputed dataset to a new CSV file
output_path = "MissForest欠損値補完後.csv"
data_imputed_df.to_csv(output_path, index=False, encoding="utf-8")

print(f"\nImputation completed. The imputed dataset has been saved as '{output_path}'.")
```
