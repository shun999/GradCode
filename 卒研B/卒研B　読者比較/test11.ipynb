```python
import numpy as np
import matplotlib.pyplot as plt

# Generate frequency points
omega = np.linspace(-np.pi, np.pi, 1024)

# Calculate frequency response
H = 0.5 / (1 - 0.5 * np.exp(-1j * omega))

# Calculate magnitude and phase
magH = np.abs(H)
phaseH = np.angle(H)

# Create subplots
plt.figure(figsize=(10, 8))

# Magnitude response plot
plt.subplot(2, 1, 1)
plt.plot(omega, magH, 'b', linewidth=1.5)
plt.xlabel('')
plt.ylabel('|G(e^{j})|')
plt.title('Magnitude Response')
plt.grid(True)

# Phase response plot
plt.subplot(2, 1, 2)
plt.plot(omega, phaseH, 'r', linewidth=1.5)
plt.xlabel('')
plt.ylabel('Phase (radians)')
plt.title('Phase Response')
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
N = 100  # Number of samples
n = np.arange(N)
omega0 = np.pi/4  # Angular frequency of input sine wave

# Input signal
x = np.sin(omega0 * n)

# Initialize output signal
y = np.zeros(N)

# Recursive calculation using difference equation
# (Note: Python is 0-indexed, so we start from index 1)
for k in range(1, N):
    y[k] = 0.5 * x[k] + 0.5 * y[k-1]

# Create plots
plt.figure(figsize=(10, 8))

# Input signal plot
plt.subplot(2, 1, 1)
plt.plot(n, x, 'b', linewidth=1.5)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('Input Signal x[n]')
plt.grid(True)

# Output signal plot
plt.subplot(2, 1, 2)
plt.plot(n, y, 'r', linewidth=1.5)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Output Signal y[n]')
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
