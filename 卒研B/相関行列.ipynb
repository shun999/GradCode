```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# file_path =  "C:/Users/thisi/OneDrive/共有ファイル/リモートデスクトップ/vif_result_test.csv"
# file_path =  "C:/Users/thisi/OneDrive/共有ファイル/リモートデスクトップ/データセット/データ最終版/1. 最終ファイルVIF処理後.csv"
# file_path =  "C:/Users/thisi/OneDrive/共有ファイル/リモートデスクトップ/データセット/データ最終版/2. 最終ファイルVIF主成分分析後.csv"
file_path =  "C:/Users/thisi/OneDrive/共有ファイル/リモートデスクトップ/データセット/データ最終版/0. 最終ファイル.csv"

data = pd.read_csv(file_path)
data = data.fillna(data.mean())
data.info()
```

```python
# データの可視化
plt.figure(figsize=(50, 40))
corr_matrix = data.corr('pearson')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```
