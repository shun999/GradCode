````python
import spacy
import ginza
import networkx as nx
import matplotlib.pyplot as plt

def analyze_dependency(file_path):
    """
    テキストファイルを読み込み、係り受け解析を行い、可視化する関数
    
    Args:
        file_path (str): 解析するテキストファイルのパス
    """
    # 日本語モデルの読み込み
    nlp = spacy.load('ja_core_news_lg')
    
    # テキストファイルを読み込む
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # 係り受け解析の実行
    doc = nlp(text)
    
    # 依存関係グラフの作成
    G = nx.DiGraph()
    
    # ノードとエッジの追加
    for token in doc:
        # 各トークンをノードとして追加
        G.add_node(token.text, pos=token.pos_)
        
        # 親ノードへのエッジを追加
        if token.dep_ != 'ROOT':
            G.add_edge(token.head.text, token.text, label=token.dep_)
    
    # グラフの可視化
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, k=1, iterations=50)
    
    # ノードの描画
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
    nx.draw_networkx_labels(G, pos)
    
    # エッジの描画
    nx.draw_networkx_edges(G, pos, edge_color='gray', arrows=True)
    
    # エッジのラベル
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    
    plt.title('係り受け解析グラフ')
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('dependency_graph.png')
    plt.close()
    
    # 解析結果の出力
    print("係り受け解析の結果:")
    for token in doc:
        print(f"{token.text} - {token.pos_} -> {token.head.text} ({token.dep_})")

# 使用例
if __name__ == "__main__":
    # テキストファイルのパスを指定
    file_path = "E:\実験\実験結果/18.文章25\文章25.txt"
    analyze_dependency(file_path)
````

````python
import spacy
import networkx as nx
import matplotlib.pyplot as plt

def analyze_dependency(file_path):
    """
    テキストファイルを読み込み、係り受け解析を行い、可視化する関数
    
    Args:
        file_path (str): 解析するテキストファイルのパス
    """
    # Ginzaの日本語モデルの読み込み
    try:
        import ja_ginza
        nlp = ja_ginza.load()
    except ImportError:
        print("Ginzaモデルがインストールされていません。")
        print("以下のコマンドでインストールしてください:")
        print("pip install ginza")
        print("python -m ginza download")
        return
    
    # テキストファイルを読み込む
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # 係り受け解析の実行
    doc = nlp(text)
    
    # 依存関係グラフの作成
    G = nx.DiGraph()
    
    # ノードとエッジの追加
    for sent in doc.sents:
        for token in sent:
            # 各トークンをノードとして追加
            G.add_node(token.text, pos=token.pos_)
            
            # 親ノードへのエッジを追加
            if token.dep_ != 'ROOT':
                G.add_edge(token.head.text, token.text, label=token.dep_)
    
    # グラフの可視化
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, k=1, iterations=50)
    
    # ノードの描画
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
    nx.draw_networkx_labels(G, pos)
    
    # エッジの描画
    nx.draw_networkx_edges(G, pos, edge_color='gray', arrows=True)
    
    # エッジのラベル
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    
    plt.title('係り受け解析グラフ')
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('dependency_graph.png')
    plt.close()
    
    # 解析結果の出力
    print("係り受け解析の結果:")
    for sent in doc.sents:
        for token in sent:
            print(f"{token.text} - {token.pos_} -> {token.head.text} ({token.dep_})")

# 使用例
if __name__ == "__main__":
    # テキストファイルのパスを指定
    file_path = "E:\実験\実験結果/18.文章25\文章25.txt" 
    analyze_dependency(file_path)


"""
依存関係のインストール手順:
```bash
pip install ginza networkx matplotlib
python -m ginza download
```

主な変更点:
- spaCyではなくGinzaの日本語モデルを直接使用
- モデルが見つからない場合のエラーハンドリングを追加
- インストール方法の案内を追加

もし引き続きエラーが発生する場合は、以下の点を確認してください：
1. Pythonのバージョンが最新であること
2. 必要なライブラリが最新版であること
3. 仮想環境を使用している場合は、正しい環境でインストールしていること

エラーの詳細や、具体的な環境情報（Python version、pip version）があれば、さらに詳細なサポートができます。"""
````
