```python
import csv

# テキストファイルの読み込み
input_file = 'C:\WorkSpace\Practice\メモ帳の行と列交換\Reading metrix.txt'
with open(input_file, 'r', encoding='utf-8') as file:
    lines = [line.strip().split('\t') for line in file]

# 行と列を入れ替える
transposed_lines = list(map(list, zip(*lines)))

# CSVファイルに出力（UTF-8 with BOM）
output_file = 'Reading metrix_sample.csv'
with open(output_file, 'w', newline='', encoding='utf-8-sig') as file:
    writer = csv.writer(file)
    writer.writerows(transposed_lines)

print(f'CSVファイル「{output_file}」に出力しました。')
```
